```python
#ONLY THIS CELL IS REQUIRED TO BE RUN.
#MODEL AND DATASET NEEDS TO BE SAVED IN THE LOCAL MACHINE AND THE DIRECTORY PATH NEEDS TO BE CHANGED ACCORDINGLY
#Real Power
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pickle

loaded_model = pickle.load(open("/content/finalized_model_SJT_tower.sav", 'rb')) #Path needs to be changed as per the lcoal machine
X = pd.read_csv("/content/Data Used for Real Power Prediction.csv") #Path needs to be changed as per the lcoal machine
X.drop(columns=['Unnamed: 0'],axis=1, inplace=True)

a = int(input("Enter the Day: "))
b = int(input("Enter the month: "))

features_day = []
features_month = []
features = []

for i in X[X['Day'] == a].columns:
  if (a==5) or (a==6) or (a==7) or (a==8) or (a==9):
    a = 4
    d = X[X['Day'] == a]['{}'.format(i)].mean()
    features_day.append(d)


  elif (a==10) or (a==11) or (a==12):
    a = 13
    d = X[X['Day'] == a]['{}'.format(i)].mean()
    features_day.append(d)

  else:
    d = X[X['Day'] == a]['{}'.format(i)].mean()
    features_day.append(d)

for i in X[X['Month'] == b].columns:
  c = X[X['Month'] == b]['{}'.format(i)].mean()
  features_month.append(c)


for j in range(0,len(features_day)):
  features.append(features_day[j] + features_month[j])

for i in range(0,len(features)):
  features[i] = features[i]/2

final_features = [features]

preds = loaded_model.predict(final_features)

print("The Real Power is: ", preds)
```

```python
import math
hours = int(input("Enter the hours: "))
mins = int(input("Enter the minutes: "))
month = int(input("Enter the Month: "))
day = int(input("Enter the Day: "))

d = X[X['Hour'] == hours][['Day','Month','Mins','PowerFactor']]
a = d[d['Mins'] == mins][['PowerFactor','Month','Day']]
c = a[a['Month'] == month][['PowerFactor', 'Day']]
b = c[c['Day'] == day]['PowerFactor'].iloc[0]

print("PowerFactor for the said date and time is: ", b)

theta = np.arccos(b)
print(theta)

sin_value = math.sin(theta)

reactive_power = (preds / b) * sin_value

print(reactive_power)
```
